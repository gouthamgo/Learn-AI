   - **Answer:** Accuracy alone can be misleading, especially with imbalanced data. 
   Check precision, recall, F1 score. The model might be predicting majority class.

2. "Your model works great on training data but poorly on production. What's wrong?"
   - **Answer:** Overfitting. Model memorized training data. Solutions: 
   regularization, more data, simpler model, better validation.

3. "How would you deploy a model that needs to handle 10K requests/second?"
   - **Answer:** Model optimization (quantization, pruning), caching, 
   load balancing, horizontal scaling, asynchronous processing.

4. "Your model's accuracy dropped from 90% to 75% after 3 months. Why?"
   - **Answer:** Data drift. Input distribution changed. Solution: 
   Monitor distributions, retrain regularly, implement online learning.
