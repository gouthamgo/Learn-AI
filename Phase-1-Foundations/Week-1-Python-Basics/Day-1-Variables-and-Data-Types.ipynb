## üì¶ What is a Variable?

Think of a variable as a **labeled box** that stores information.

In real life:
- A box labeled "Toys" contains toys
- A box labeled "Books" contains books

In Python:
- A variable named `age` can store a number like 25
- A variable named `name` can store text like "Alice"

Let's see this in action! üëá

```python
# This is a comment - Python ignores it. Use # to write notes!

# Creating our first variable
name = "AI Learner"  # ‚Üê The box is labeled 'name'
                     # ‚Üê It contains the text "AI Learner"

# Let's see what's inside the box
print(name)  # print() shows us the content
```

**üéâ You just created your first variable!**

**What happened?**
1. We created a box labeled `name`
2. We put "AI Learner" inside it
3. We used `print()` to display what's inside

## üî¢ Data Types: Different Kinds of Data

Python has different **types** of data, just like we have different types of boxes:

| Type | Python Name | Example | Used For |
|------|-------------|---------|----------|
| Text | `str` (string) | `"Hello"` | Names, messages, text data |
| Whole Number | `int` (integer) | `42` | Counting, IDs, ages |
| Decimal Number | `float` | `3.14` | Measurements, prices, predictions |
| True/False | `bool` (boolean) | `True` or `False` | Yes/no decisions |

Let's explore each one! üëá

### 1Ô∏è‚É£ Strings (Text Data)

```python
# Strings = Text (always in quotes)

greeting = "Hello, AI World!"  # ‚Üê Text in double quotes
message = 'Python is awesome'   # ‚Üê Or single quotes (both work!)

print(greeting)
print(message)

# Check the type
print(type(greeting))  # type() tells us what kind of data it is
```

**üß† AI Connection:**
- Strings store text data for chatbots, sentiment analysis
- Movie reviews, tweets, emails = all strings!
- NLP (Natural Language Processing) works with strings

### 2Ô∏è‚É£ Integers (Whole Numbers)

```python
# Integers = Whole numbers (no quotes!)

age = 25              # ‚Üê A positive number
temperature = -5      # ‚Üê Can be negative
pixels = 1920         # ‚Üê Image width in pixels

print(age)
print(temperature)
print(type(pixels))   # Check the type
```

**üß† AI Connection:**
- Image dimensions: 1920x1080 pixels
- Number of training examples: 50,000 images
- Epochs (training rounds): 100

### 3Ô∏è‚É£ Floats (Decimal Numbers)

```python
# Floats = Numbers with decimals

price = 19.99          # ‚Üê Money amounts
accuracy = 0.95        # ‚Üê ML model accuracy (95%)
pi = 3.14159          # ‚Üê Mathematical constants

print(f"Model accuracy: {accuracy}")  # f-string to format output
print(type(accuracy))
```

**üß† AI Connection:**
- Model accuracy: 0.95 (95% correct)
- Probabilities: 0.87 (87% chance of rain)
- Learning rate: 0.001 (how fast model learns)

### 4Ô∏è‚É£ Booleans (True/False)

```python
# Booleans = True or False (only these two values!)

is_raining = True      # ‚Üê Note: Capital T
is_sunny = False       # ‚Üê Capital F
model_trained = True

print(is_raining)
print(type(is_raining))
```

**üß† AI Connection:**
- Email is spam: True or False?
- Image contains a cat: True or False?
- Customer will buy: True or False?

## üîÑ Variables Can Change!

```python
# Variables are called "variable" because they can CHANGE!

score = 0         # Start with 0
print(f"Score: {score}")

score = 10        # Now it's 10
print(f"Score: {score}")

score = score + 5  # Add 5 to current score
print(f"Score: {score}")
```
