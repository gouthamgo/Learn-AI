## ðŸ“‹ What is a List?

Yesterday we learned about variables that store ONE thing.

But what if you have:
- 1000 student names?
- 50,000 product prices?
- 1 million images?

**Lists** let you store MANY items in one variable!

Think of it as a shopping list:
```
Shopping List:
1. Milk
2. Eggs
3. Bread
4. Apples
```

````python
# Creating a list (use square brackets [])

shopping_list = ["Milk", "Eggs", "Bread", "Apples"]

print(shopping_list)
````

**ðŸŽ‰ You just created a list!**

**Parts of a list:**
- `[` and `]` = Square brackets (wrap the list)
- Items separated by commas `,`
- Can contain any data type!

````python
# Lists can hold different types of data

numbers = [1, 2, 3, 4, 5]                    # List of integers
prices = [19.99, 29.99, 9.99]                # List of floats
names = ["Alice", "Bob", "Charlie"]          # List of strings
mixed = ["AI", 2024, 3.14, True]             # Mixed types!

print(numbers)
print(prices)
print(names)
print(mixed)
````

## ðŸ” Accessing List Items

Each item in a list has an **index** (position number).

**âš ï¸ IMPORTANT:** Python starts counting at 0 (not 1!)

```
fruits = ["Apple", "Banana", "Orange"]
          â†‘        â†‘         â†‘
        index 0  index 1   index 2
```

````python
# Accessing items by index

fruits = ["Apple", "Banana", "Orange", "Mango"]

print(fruits[0])   # First item (index 0)
print(fruits[1])   # Second item (index 1)
print(fruits[2])   # Third item (index 2)
print(fruits[-1])  # Last item (negative index!)
````

## âž• Modifying Lists

````python
# Lists can be changed!

todos = ["Study", "Exercise", "Code"]
print("Before:", todos)

# Add an item to the end
todos.append("Sleep")
print("After append:", todos)

# Remove an item
todos.remove("Exercise")
print("After remove:", todos)

# Change an item
todos[0] = "Study AI"
print("After change:", todos)

# How many items?
print(f"Total tasks: {len(todos)}")
````

## ðŸ”„ Loops: Processing Every Item

Imagine you have 1000 student names and want to print each one.

**Without loops:** You'd write 1000 print statements! ðŸ˜±

**With loops:** Just 2 lines of code! ðŸŽ‰

**A loop** repeats code automatically for each item.

````python
# The FOR loop - most common in AI/ML!

fruits = ["Apple", "Banana", "Orange"]

# This loop runs 3 times (once for each fruit)
for fruit in fruits:
    print(f"I like {fruit}")
    
print("Done!")
````

**What's happening?**
1. Python takes the first item ("Apple") and puts it in `fruit`
2. Runs the indented code
3. Takes the next item ("Banana"), repeats
4. Takes the next item ("Orange"), repeats
5. No more items? Stop the loop!

**âš ï¸ IMPORTANT:** Notice the **indentation** (spaces before `print`). That's how Python knows what's inside the loop!

````python
# Real AI example: Processing accuracy scores

model_accuracies = [0.85, 0.92, 0.88, 0.95, 0.90]

print("Model Performance Report:")
print("=" * 30)

for accuracy in model_accuracies:
    percentage = accuracy * 100
    print(f"Accuracy: {percentage}%")
````

## ðŸ”¢ Loops with Numbers: range()

Sometimes you want to repeat something N times:

````python
# Print numbers 0 to 4
for i in range(5):
    print(i)
````

````python
# Print numbers 1 to 10
for i in range(1, 11):  # Start at 1, stop before 11
    print(i)
````

````python
# Real AI example: Training epochs

print("Training AI Model...")
print()

for epoch in range(1, 6):  # Train for 5 epochs
    print(f"Epoch {epoch}/5 - Training...")
    
print("\nâœ… Training complete!")
````

## ðŸš€ Real-World AI Example: Image Dataset

Let's simulate processing images for an AI model:

````python
# Simulating an image dataset
image_files = [
    "cat_001.jpg",
    "dog_045.jpg",
    "cat_002.jpg",
    "bird_033.jpg",
    "dog_046.jpg"
]

# Count each category
cat_count = 0
dog_count = 0
bird_count = 0

print("Processing images...\n")

for image in image_files:
    print(f"Loading: {image}")
    
    # Check what type of image it is
    if "cat" in image:
        cat_count += 1
    elif "dog" in image:
        dog_count += 1
    elif "bird" in image:
        bird_count += 1

# Print summary
print("\n" + "=" * 40)
print("DATASET SUMMARY")
print("=" * 40)
print(f"Total images: {len(image_files)}")
print(f"Cats: {cat_count}")
print(f"Dogs: {dog_count}")
print(f"Birds: {bird_count}")
````

**ðŸ§  This is EXACTLY what AI models do!**
- Load each image from a list
- Process it (resize, normalize, classify)
- Count categories
- Generate reports
